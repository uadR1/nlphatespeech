```python
import pandas as pd
import numpy as np

from sklearn.metrics import confusion_matrix
import seaborn as sns

import matplotlib.pyplot as plt
```

# Combined the Ground Truth & Survey Result

```python
df_ground_truth = pd.read_csv('exported/ground_truth.csv', index_col='sheet_index')
df_ground_truth['y_true']=0
df_ground_truth.loc[df_ground_truth['explicit_hate']==1,'y_true']=3
df_ground_truth.loc[df_ground_truth['implicit_hate']==1,'y_true']=2
df_ground_truth.loc[df_ground_truth['offensive']==1,'y_true']=1


df_ground_truth[10:200]
```

```python
df_result = pd.read_csv('exported/survey_result_v2.csv', index_col='sheet_index')
df_result['y_pred']=0
df_result.loc[df_result['explicit_hate_prediction']==1,'y_pred']=3
df_result.loc[df_result['implicit_hate_prediction']==1,'y_pred']=2
df_result.loc[df_result['offensive_prediction']==1,'y_pred']=1

df_result[10:100]
```

```python
print(df_ground_truth.shape)
print(df_result.shape)
```

```python

# merge ground truth & survey result
df_survey = pd.merge(df_ground_truth, df_result, how='left', left_index=True, right_index=True)

df_survey
```

```python
y_test = list(df_survey['y_true'])
y_prediction = list(df_survey['y_pred'])
```

```python
from sklearn import metrics
```

```python
# MULTI-CLASS CLASSIFICATION SCORES
# Accuracy Score 
accuracy = metrics.accuracy_score(y_test, y_prediction)
print('Accuracy: ',accuracy)

# Precision Score Weighted
precision = metrics.precision_score(y_test, y_prediction, average='weighted')
print('Precision: ',precision)

# Recall Score Weighted
recall = metrics.recall_score(y_test, y_prediction, average='weighted')
print('Recall: ',recall)

# F1 Score Weighted
f1 = metrics.f1_score(y_test, y_prediction, average='weighted')
print('F1: ',f1)
```
