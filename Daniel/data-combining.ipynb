# Combined dataset

#### DS 1 implicit

```python
df_explicit_implicit = pd.read_csv('dataset/implicit-hate-corpus/implicit_hate_v1_stg1_posts.tsv',delimiter='\t')
df_explicit_implicit.rename(columns={'post':'text'}, inplace=True)


df1_explicit = df_explicit_implicit[df_explicit_implicit['class']=='explicit_hate']
df1_implicit = df_explicit_implicit[df_explicit_implicit['class']=='implicit_hate']
df1_nohate = df_explicit_implicit[df_explicit_implicit['class']=='not_hate']
df1_nohate['class']='not_hate'


print(df1_explicit.shape)
print(df1_implicit.shape)
print(df1_nohate.shape)
```

#### DS 2 hate vs offensive

```python
df_hate_n_offensive = pd.read_csv('dataset/hate-speech-and-offensive-language/labeled_data.csv')
df_hate_n_offensive = df_hate_n_offensive[['tweet','class']]
df_hate_n_offensive.rename(columns={'tweet':'text'}, inplace=True)

df2_hate = df_hate_n_offensive[df_hate_n_offensive['class']==0]
df2_hate['class']='explicit_hate'
df2_offensive = df_hate_n_offensive[df_hate_n_offensive['class']==1]
df2_offensive['class'] = 'offensive'
df2_nohate = df_hate_n_offensive[df_hate_n_offensive['class']==2]
df2_nohate['class'] = 'not_hate'
print(df2_hate.shape)
print(df2_offensive.shape)
print(df2_nohate.shape)
```

```python
df2_nohate
```

#### DS 3 neutral

```python
df_neutral = pd.read_csv('dataset/neutral-dataset/GabHateCorpus_annotations.tsv', delimiter='\t')

df_neutral_filtered = df_neutral.groupby(['ID'], as_index=False)[['Hate', 'HD','CV','VO']].mean() 
df_neutral_filtered['no_hate_at_all'] = df_neutral_filtered['Hate'] + df_neutral_filtered['HD'] + df_neutral_filtered['CV'] + df_neutral_filtered['VO'] 
df_neutral_filtered = df_neutral_filtered[df_neutral_filtered['no_hate_at_all']==0]

df_neutral_drop_dup = df_neutral.drop_duplicates(subset=['Text'], keep='first')
df_neutral = pd.merge(df_neutral_filtered, df_neutral_drop_dup, how='left', left_on='ID', right_on='ID')
df_neutral = df_neutral[['Text','no_hate_at_all']]
df_neutral['no_hate_at_all'] = 'not_hate'

df_neutral.rename(columns={'Text':'text','no_hate_at_all':'class'}, inplace=True)
df3_nohate = df_neutral
print(df_neutral.shape)
df_neutral.head()
```

#### COMBINING

```python
import random
random.seed(1)

def list_of_random_index(df, num_samples):
    return random.sample(list(range(0,len(df))), num_samples)

# DS 1
df1_explicit_1070 = df1_explicit.iloc[list_of_random_index(df1_explicit, 1070)]
df1_implicit_2500 = df1_implicit.iloc[list_of_random_index(df1_implicit, 2500)]
df1_nohate_2500 = df1_nohate.iloc[list_of_random_index(df1_nohate, 2500)]

# DS 2
df2_hate_1430 = df2_hate
df2_offensive_2500 = df2_offensive.iloc[list_of_random_index(df2_offensive, 2500)]
df2_nohate_2500 = df2_nohate.iloc[list_of_random_index(df2_nohate, 2500)]

# DS 3
df3_nohate_2500 = df3_nohate.iloc[list_of_random_index(df3_nohate, 2500)]


# Combined explicit_hate & no_hate
df_explicit_combined_2500 = pd.concat([df1_explicit_1070, df2_hate_1430])
df_nohate_combined_7500 = pd.concat([df1_nohate_2500, df2_nohate_2500, df3_nohate_2500])

# Combined all for binary classification
df_combined_binary = pd.concat([df_explicit_combined_2500, df1_implicit_2500, df2_offensive_2500, df_nohate_combined_7500 ])

print('explicite_hate \t', df_explicit_combined_2500.shape)
print('implicit_hate \t', df1_implicit_2500.shape)
print('offensive \t', df2_offensive_2500.shape)
print('no_hate \t', df_nohate_combined_7500.shape)

print('combined dataset for binary classification \t', df_combined_binary.shape)
```

```python
df_combined_binary.to_csv('exported/df_combined_binary.csv')
```

# Human Baseline Experiment Setup

```python
import random
random.seed(1)

def list_of_random_index(df, num_samples):
    return random.sample(list(range(0,len(df))), num_samples)

# DS 1
df_survey_explicit_150 = df_explicit_combined_2500.iloc[list_of_random_index(df_explicit_combined_2500, 150)]
df_survey_implicit_150 = df1_implicit_2500.iloc[list_of_random_index(df1_implicit_2500, 150)]
df_survey_offensive_150 = df2_offensive_2500.iloc[list_of_random_index(df2_offensive_2500, 150)]
df_survey_nohate_450 = df_nohate_combined_7500.iloc[list_of_random_index(df_nohate_combined_7500, 450)]

# export the survey file
df_survey = pd.concat([df_survey_explicit_150, df_survey_implicit_150, df_survey_offensive_150, df_survey_nohate_450])
df_survey = pd.concat([df_survey, pd.get_dummies(df_survey['class'])], axis=1)

df_survey.to_csv('exported/df_survey.csv')
```

```python
df_survey['class'].value_counts()
```

```python
# Shuffle before sending to the annotators

import random
random.seed(1)

def list_of_random_index(from_ ,to_):
    return random.sample(list(range(from_,to_)), (to_ - from_))

index_survey_shuffled= list_of_random_index(0,900)

df_survey_shuffled = df_survey.iloc[index_survey_shuffled,:]

df_survey_shuffled.to_csv('exported/df_survey_shuffled.csv')
```
